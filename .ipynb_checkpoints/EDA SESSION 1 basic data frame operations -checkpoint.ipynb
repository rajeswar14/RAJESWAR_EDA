# creating data frames using list

```python
name=['Ramesh','Suresh','Sathish']
age=[30,35,40]
name,age
```

# step-1 create data frame

```python
pd.DataFrame() # to make the data frame
```

# step2
$provide$ $data$

```python
pd.DataFrame(zip(name,age)) 
```

# step3
$provide$ $columns$

```python
# provide columns
data=zip(name,age)
cols=['name','age']
pd.DataFrame(zip(name,age),columns=['NAME','AGE'])
```

```python
data=zip(name,age)
cols=['name','age']
pd.DataFrame(data,columns=cols)
```

# step 4
$provide$ $index$

```python
data=zip(name,age)
cols=['name','age']
ind=['A','B','C']
pd.DataFrame(data,columns=cols,index=ind)
```

# step 5 
$add$ $new$ $columns$

```python
name=['Ramesh','Suresh','Sathish']
age=[30,35,40]
data=zip(name,age)
cols=['name','age']
ind=['A','B','C']
df=pd.DataFrame(data,columns=cols,index=ind)
df
```

- if u want to add new columns
- df['new colums name']
- you need to have a list which is having some elements
- city_names=['hyd'.'blr','Chennai']
- df["city"]=city_names


```python
city_names=['hyd','banglore']
df["city"]=city_names
df
```

```python
city_names=['hyd','bangalore','chennai']
df["city"]=city_names
df
```

# step-6**
$update$ $the$ $existing$ $column$

- if you want to create new column or update the old column
- both are same way

```python
df['name']=['RAJESH','KAUSHIK','MADHU']
df
```

$drop$ $the$ $column$
 - in order to drop the  column
 - we need to use drop method
 - it takes 3 parameters
 - drop column or row
 - mention the column name
 - axis
 

```python
df.drop('city',
        axis=1,
       inplace=True)
```

```python
name=['Ramesh','Suresh','Sathish']
age=[30,35,40]
data=zip(name,age)
cols=['name','age']
ind=['A','B','C']
df=pd.DataFrame(data,columns=cols,index=ind)
city_names=['hyd','bangalore','chennai']
df["city"]=city_names #column_name
df.drop('city',#column_name
       axis=1,# column
       inplace=True)# over write the same
df
```

$step$ $8$

- drop the rows

```python
df.drop('C',#column_name
       axis=0,# column
       inplace=True)# over write the same
df
```

# step 9

$save$ $the$ $data$ $frame$

```python
df.to_csv("dataframe.csv")#to csv
#while copying index consider as extra column so unnamed column is showing below
df.to_excel("dataframe.xlsx") #to excel format
```
